```python
from langchain_community.document_loaders import PyPDFDirectoryLoader, UnstructuredPowerPointLoader, CSVLoader, JSONLoader, UnstructuredExcelLoader, Docx2txtLoader
from pptx import Presentation
import json
from pathlib import Path

def load_data_from_files(pdf_dir, ppt_file, csv_file, json_file, docx_file, text_file, xlsx_file):
    # Load data from PDF directory
    pdf_loader = PyPDFDirectoryLoader(pdf_dir)
    pdf_data = pdf_loader.load()

    # Load data from PowerPoint presentation
    powerpoint_loader = UnstructuredPowerPointLoader(ppt_file)
    powerpoint_data = powerpoint_loader.load()

    # Load data from CSV file
    csv_loader = CSVLoader(file_path=csv_file)
    csv_data = csv_loader.load()

    # Load data from JSON file
    json_data = json.loads(Path(json_file).read_text())

    # Load data from Excel file
    #excel_loader = UnstructuredExcelLoader(excel_file, mode="elements")
    #excel_data = excel_loader.load()

    # Load data from DOCX file
    docx_loader = Docx2txtLoader(docx_file)
    docx_data = docx_loader.load()

    #load data from the text file
    from langchain.document_loaders import UnstructuredFileLoader
    text_loadeer=UnstructuredFileLoader(text_file)
    text_data=text_loadeer.load()
    text_data

    #loading data from the xlxs file
    from langchain_community.document_loaders import UnstructuredExcelLoader
    xlsx_loader = UnstructuredExcelLoader(xlsx_file, mode="elements")
    xlsx_data = xlsx_loader.load()
    xlsx_data[0]

    html_content=str(xlsx_data)
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')
    rows = soup.find('tbody').find_all('tr')
    for row in rows:
      cells = row.find_all('td')
      excel_Data = [cell.get_text() for cell in cells]
      print(excel_Data)





    return pdf_data, powerpoint_data, csv_data, json_data,  docx_data, text_data, xlsx_data

# Replace file paths with your actual file paths
pdf_dir = "/content/pdf data"
ppt_file = "/content/pdf data/Nimble Property .pptx"
csv_file = "/content/pdf data/Book1.csv"
json_file = "/content/pdf data/intents.json"
#excel_file = "/content/pdf data/Book1.xlsx"
docx_file = "/content/pdf data/CHATBOT DOCUEMENTATION.docx"
text_file = "/content/pdf data/elon_4.txt"
xlsx_file = "/content/pdf data/Book1.xlsx"
# Load data from files
pdf_data, powerpoint_data, csv_data, json_data, docx_data, text_data, excel_Data = load_data_from_files(pdf_dir, ppt_file, csv_file, json_file, docx_file, text_file, xlsx_file)

# Print or process the loaded data as needed
print("PDF Data:", pdf_data)
print("PowerPoint Data:", powerpoint_data)
print("CSV Data:", csv_data)
print("JSON Data:", json_data)
#print("Excel Data:", excel_data)
print("DOCX Data:", docx_data)
print("text_Data", text_data)
print("xlsx_Data", excel_Data)
```
